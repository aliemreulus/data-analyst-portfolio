```python
import pandas as pd
import numpy as np

np.random.seed(10)
orders = pd.DataFrame({
    'order_id':    range(1, 201),
    'customer_id': np.random.randint(1, 51, 200),
    'product':     np.random.choice(['Laptop','Phone','Tablet','Monitor','Keyboard'], 200),
    'amount':      np.random.uniform(50, 2000, 200).round(2),
    'date':        pd.date_range('2024-01-01', periods=200, freq='2D'),
})
orders.head()
```

## Query 1 – Top 10 customers by revenue

```python
orders.groupby('customer_id')['amount'].sum().sort_values(ascending=False).head(10)
```

## Query 2 – Monthly revenue trend

```python
monthly = orders.resample('ME', on='date')['amount'].sum().reset_index()
monthly.columns = ['month', 'revenue']
monthly['month'] = monthly['month'].dt.strftime('%Y-%m')
print(monthly.to_string(index=False))
```

## Query 3 – Best-selling product

```python
orders.groupby('product')['amount'].agg(['sum','count']).sort_values('sum', ascending=False)
```
